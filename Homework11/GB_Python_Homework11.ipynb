```python
from sympy import *
from sympy.plotting import plot

init_printing()
```

```python
x = Symbol('x')
```

```python
fx = 5*x**2 + 10*x - 30
```

```python
fx
```

```python
#    4. Построить график
plot(fx)
```

```python
#    1. Определить корни
solve(fx)
```

```python
#    5. Вычислить вершину
p = diff(fx)
p
```

```python
solve(p)
```

```python
fx.subs(x, -1)
```

```python
#    3. Найти интервалы, на которых функция убывает
# Функция убывает на промежутке x = (-infinity; -1)
```

```python
#    6. Определить промежутки, на котором f > 0
solve_univariate_inequality(fx > 0, x)
```
